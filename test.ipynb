```python
import os, json, requests

base = os.getenv('COEVA_OPENAI_BASE_URL', 'https://api.openai.com').rstrip('/') + '/v1/chat/completions'
model = os.getenv('COEVA_OPENAI_MODEL', 'gpt-4o-mini')
key   = os.getenv('OPENAI_API_KEY', '')

print("POST", base, "model=", model)
r = requests.post(base,
    headers={'Authorization': f'Bearer {key}', 'Content-Type': 'application/json'},
    json={'model': model, 'messages':[{'role':'user','content':'ping'}], 'max_tokens': 8})
print("status:", r.status_code)
print("body:", r.text[:500])
```
